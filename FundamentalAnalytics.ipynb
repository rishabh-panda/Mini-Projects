```
#Sine Signal

t=np.arange(-3.0,3.0,0.05)
v=np.sin(2*np.pi*t)
plt.figure(figsize=(8,5))
plt.plot(t,v)
plt.title("Sine Signal",fontsize=15)
plt.ylabel("Amplitude",fontsize=12)
plt.xlabel("Time",fontsize=12)
plt.show()
```

```
#Cos Signal

t=np.arange(-3.0,3.0,0.05)
v=np.cos(2*np.pi*t)
plt.figure(figsize=(8,5))
plt.plot(t,v)
plt.title("Cosine Signal",fontsize=15)
plt.ylabel("Amplitude",fontsize=12)
plt.xlabel("Time",fontsize=12)
plt.show()
```

```
#Unit Step Signal

ar1=np.zeros(10,dtype=int)
ar2=np.ones(11,dtype=int)
ar3=np.concatenate((ar1,ar2))
print(ar3)
t=np.arange(-10.0,11.0,1.0)
plt.figure(figsize=(8,5))
plt.stem(t,ar3,use_line_collection=True)
plt.title("Unit Step Signal",fontsize=15)
plt.ylabel("Amplitude",fontsize=12)
plt.xlabel("Time",fontsize=12)
plt.show()
```

```
#Ramp Signal

ar1=np.zeros(11,dtype=int)
ar2= [i for i in range(1,11)]
ar3=np.concatenate((ar1,ar2))
print(ar3)
t=np.arange(-10.0,11.0,1.0)
plt.figure(figsize=(8,5))
plt.stem(t,ar3,use_line_collection=True)
plt.title("Ramp Signal",fontsize=15)
plt.ylabel("Amplitude",fontsize=12)
plt.xlabel("Time",fontsize=12)
plt.show()
```

```
#Unit Impulse Signal

ar1=np.zeros(10,dtype=int)
ar2= [1]
ar3=np.zeros(10,dtype=int)
ar4=np.concatenate((ar1,ar2,ar3))
print(ar4)
t=np.arange(-10.0,11.0,1.0)
plt.figure(figsize=(8,5))
plt.stem(t,ar4,use_line_collection=True)
plt.title("Unit Impulse Signal",fontsize=15)
plt.ylabel("Amplitude",fontsize=12)
plt.xlabel("Time",fontsize=12)
plt.show()
```

```
#Sawtooth Signal

from scipy import signal
import matplotlib.pyplot as plt
t = np.linspace(0, 1, 500)
plt.figure(figsize=(8,5))
plt.plot(t, signal.sawtooth(2 * np.pi * 5 * t))
plt.title("Sawtooth Signal",fontsize=15)
plt.ylabel("Amplitude",fontsize=12)
plt.xlabel("Time",fontsize=12)
```
